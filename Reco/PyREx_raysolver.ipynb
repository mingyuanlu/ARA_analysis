```python
import numpy as np
import matplotlib.pyplot as plt
import pyrex
```

```python
# rt = pyrex.RayTracer(from_point=(0, 0, -1250), to_point=(2400, 0, -179.93))
rt = pyrex.RayTracer(from_point=(0, 0, -495.0), to_point=(1566.8, 0, -2.9e+02))
```

```python
path = rt.solutions[0]
path2 = rt.solutions[1]
```

```python
90-np.degrees(path.theta0)
```

```python
np.degrees(np.arctan(2400/(-1200+179.93)))+90
```

```python
theta_d=-np.degrees(np.arctan(path.received_direction[2]/path.received_direction[0]))
```

```python
theta_r=-np.degrees(np.arctan(path2.received_direction[2]/path2.received_direction[0]))
```

```python
plt.plot(path.coordinates[0],path.coordinates[2], label="Direct ray ($\\theta_d =%0.1f^\circ$)"% theta_d)
plt.plot(path2.coordinates[0],path2.coordinates[2], label="Reflected ray ($\\theta_r =%0.1f^\circ$)"% theta_r)
# plt.hlines(0,0,2500)
# plt.plot(2400, -179.93,"o", label="A2")
# plt.plot(0, -1200, "^", label="SPIceCore")
plt.legend()
plt.ylabel("Depth [m]",fontsize=12)
plt.xlabel("Distance [m]",fontsize=12)
# plt.annotate('A2', xy=(2400, -179.93),  xycoords='data',
#             xytext=(1, 0.5), textcoords='axes fraction',
#             arrowprops=dict(facecolor='black', shrink=0.1, width=2),
#             horizontalalignment='right', verticalalignment='top',
#             )
plt.title("Ray solutions between A2 and SPIceCore", fontsize=14)
plt.tight_layout()
plt.savefig("./A2_SPIceCore.png", dpi=250)
```
